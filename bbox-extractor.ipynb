```python
import numpy as np
import os

# image processing
from skimage import color, exposure, transform, io
from PIL import Image

# viz
import matplotlib.pyplot as plt
```

```python
IMG_PATH = 'images'
LABEL_PATH = 'labels'
OUTPUT = 'outputs'
LABEL = 0
```

```python
for image in os.listdir(IMG_PATH):
    try:
        ann = np.loadtxt(os.path.join(LABEL_PATH, image.replace(image[-4:], '.txt'))).tolist()
    

        # Store the detected objects
        img = io.imread(os.path.join(IMG_PATH, image))
        h, w, c = img.shape
        i = 0
        if ann != []:
            # multiple objects
            if type(ann[0]) == list:
                for obj in ann:

                    # only extract class 0
                    if int(obj[0]) == LABEL:
                        try:
                            x = int(obj[1]*w)
                            y = int(obj[2]*h)
                            width = int(obj[3]*w)
                            height = int(obj[4]*h)
                            x = int(x - width*.5)
                            y = int(y - height*.5)

                            obj_img =  img[y:y+height, x:x+width]
                            obj_img = Image.fromarray(obj_img)

                            image = f'{i}+' + image
                            obj_img.save(os.path.join(OUTPUT, image))
                            i += 1
                        except ValueError:
                            pass

            # single object            
            else:

                # only extract class 0
                if int(ann[0]) == LABEL:
                    x = int(ann[1]*w)
                    y = int(ann[2]*h)
                    width = int(ann[3]*w)
                    height = int(ann[4]*h)
                    x = int(x - width*.5)
                    y = int(y - height*.5)

                    obj_img =  img[y:y+height, x:x+width]
                    obj_img = Image.fromarray(obj_img)

                    image = f'{i}+' + image
                    obj_img.save(os.path.join(OUTPUT, image))
                    i += 1
    except OSError:
        pass
```
